### Objetivo del dataset
El dataset con el que trabajaré arroja un conjunto de datos apreciables para el análisis de las ventas de un hipermercado duranto el preíodo 2019 hasta 2022. Aquí se podrá valorar las categorías y productos más requeridos, pero en contra partida visualizaremos aquellos que fueron menos solicitados por el cliente para proyectar como reforzar esas ventas y promover mayor circulación. También tendremos en cuenta los países que más consumen, y como redoblar la apuesta para motivar aquellos países que aún no predominan en este registro.
Trataremos de comprender si los beneficios tuvieron influencia para que genere mas ventas de productos especificos y entender si debemos cambiar de estrategia en otros productos que no fueron tanto.

### Diccionario del dataset
Irder_id (ID del pedido), Order_date (fecha del pedido), Ship_date (fecha de envío) : detalles de identificación y sincronización de cada transacción.

Customer (Cliente), Segment (Segmento) : Información sobre el comprador y el segmento de mercado (por ejemplo, Consumidor, Oficina Central).

Manufactory (Manufactura), Product_name (Nombre del producto), Category (Categoría), Subcategory (Subcategoría) : Detalles sobre el origen, tipo y clasificación del producto

Region (Región), City (Ciudad), State (Estado), Country (País), Zip (Código Postal) : Información geográfica sobre la ubicación de la transacción.

Sales (Ventas), Quantity (cantidad), Discount (descuento), Profit (beneficio), Profit_margin (margen de beneficio) : cifras financieras que proporcionan información sobre el impacto económico de la transacción.

Grafico temporal - Lineas de ventas a lo largo del tiempo

```python
sales_over_time = data.groupby('order_date')['sales'].sum().reset_index()

plt.figure(figsize=(30, 10))
plt.plot(sales_over_time['order_date'], sales_over_time['sales'])
plt.title('Ventas a lo largo del tiempo')
plt.xlabel('Fecha de Orden')
plt.ylabel('Ventas Totales')
plt.xticks(rotation=45)
plt.show()
```

Debido a que las fechas de las ventas son por year/month/day, intenté filtrarlas ya sea por un rango especifico o por year o month. En este caso lo filtrare por un rango de fechas especifico, pero no arroja una visualización. Dejo un de los ejemplos.

```python
# Filtro de datos por rango fechas
data_filtered = data[data["order_date"] >= "2019-01-01"]
data_filtered = data_filtered[data_filtered["order_date"] <= "2022-12-31"]

# Gagrupar por order date y sum las ventas Sales
sales_over_time = data_filtered.groupby("order_date")["sales"].sum().reset_index()

# Crear el grafico de lineas
plt.figure(figsize=(30, 10))
plt.plot(sales_over_time["order_date"], sales_over_time["sales"])
plt.title("Sales over Time")
plt.xlabel("Order Date")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

Grafico de torta con los porcentajes de los paises registrados en el dataset que realizaron compras

```python
country_counts = data['country'].value_counts()
```

```python
threshold = 0.05

total_counts = country_counts.sum()

frequent_countries = country_counts[country_counts / total_counts >= threshold]
other_countries = country_counts[country_counts / total_counts < threshold]

other_countries_sum = other_countries.sum()
frequent_countries['Otros'] = other_countries_sum

print(frequent_countries)
```

Teniendo en cuenta que las ventas se realizaron solo a personas de Estados Unidos, evaluaremos los porcentajes de las ciudades.

```python
city_counts = data['city'].value_counts()
```

```python
threshold = 0.01

total_counts = city_counts.sum()

frequent_cities = city_counts[city_counts / total_counts >= threshold]
other_cities = city_counts[city_counts / total_counts < threshold]

other_cities_sum = other_cities.sum()
frequent_cities['Otros'] = other_cities_sum

print(frequent_cities)
```

Dado que las ciudades son varias, se muestra los mayores porcentajes de las ciudades que más comprano y las ciudades menos frecuentes se agruparon en Otros.

```python
total_counts = city_counts.sum()
city_percentages = (city_counts / total_counts) * 100
city_percentage_data = city_percentages.reset_index()
city_percentage_data.columns = ['city', 'percentage']
print(city_percentage_data)
```

Grafico de barras que indican las mayores compras según la Categoría

```python
import matplotlib.pyplot as plt
# Create a new DataFrame with the desired columns
city_category_sales = data.groupby(['city', 'category'])['sales'].sum().reset_index()

# Filter the DataFrame for the top 5 cities with the highest total sales
top_5_cities = city_category_sales.groupby('city')['sales'].sum().nlargest(5).index
city_category_sales_top_5 = city_category_sales[city_category_sales['city'].isin(top_5_cities)]

# Create the bar chart
plt.figure(figsize=(10, 6), facecolor='black')
sns.barplot(x='city', y='sales', hue='category', data=city_category_sales_top_5)
plt.title('Top 5 Cities with Highest Sales by Category', color='white')
plt.xlabel('City', color='white')
plt.ylabel('Total Sales', color='white')
plt.gca().set_facecolor('black')
plt.gca().tick_params(axis='x', colors='white')
plt.gca().tick_params(axis='y', colors='white')
plt.legend(facecolor='black', edgecolor='white', labelcolor='white', title='Category', title_fontsize='13', fontsize='11')
plt.show()
```

- Como resultado tenemos a New York con un volúmen importante de compras en Tecnología (Technology), Los Angeles también arroja un resultado alto en tencnología. Luego la otra categoría más vendida se trata de Muebles (Furniture) tanto en New york como en Los Angeles.
